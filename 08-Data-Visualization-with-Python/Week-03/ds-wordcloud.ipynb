```python
from wordcloud import WordCloud, STOPWORDS
from pathlib import Path
import matplotlib.pyplot as plt
import json

top = Path('.').resolve().parents[1]
print(top)
```

```python
stopwords = ['will', 'br', 'b', 'p', 'using', 'use', 'print', 'let', 'columns', 'function']
for word in stopwords:
    STOPWORDS.add(word)

words = ""
for file in top.rglob('*.ipynb'):
    data = json.loads(file.read_text())
    for cell in data['cells']:
        words += "".join(cell['source']).replace("\n", " ")

wc = WordCloud(
    normalize_plurals=False,
)
wc.generate(words)
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
